```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../documents/aliexpress laptops.csv')
```

```python
# 1) Define your bins and labels
bins = [0, 1000, 3000, 5000, 7000,  10000, 20000, np.inf]            
labels = ["0–1000", "1001–3000", "3001-5000", "5001-7000",
          "7001-10000", "10001-20000", "20000+"]

# 2) Create a new column categorizing each price into a bin
df["price_range"] = pd.cut(df["price"], bins=bins, labels=labels, right=True)
grouped = df.groupby("price_range",  observed=True).size()
```

```python
# 2. Create an improved bar chart
fig, ax = plt.subplots()
ax.bar(grouped.index.astype(str), grouped.values)

# 3. Enhance readability
ax.set_title('Number of Products by Price Range')
ax.set_xlabel('Price Range (MXN)')
ax.set_ylabel('Count of Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
output_path = '../media/price_by_range.png'
fig.savefig(output_path, dpi=300)
```
